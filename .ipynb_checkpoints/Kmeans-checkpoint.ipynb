```r
df = read.csv('data/ageinc_g.csv', row.names='X')
head(df)
```

```r
tail(df)
summary(df)
```

```r
plot(x = df$age, y =df$income, xlab = 'age', ylab='income', main = 'income with age',
     col = 'blue', pch=15)
```

### Comments:
- 4 clusters can be realized by intuition

### Buil Model

```r
mydata = df
wss = (nrow(mydata)-1)*sum(apply(mydata,2,var))
for (i in 1:10) wss[i] = sum(kmeans(mydata[,1:2],
                                   centers = i)$withinss)
plot(1:10, wss, type='b', xlab='Number of clusters',
    ylab='within groups sum of squars')
```

#### with Elbow method, 4-5-6 clusters are equals... 

### Clustering

```r
set.seed(42)
datacluster = kmeans(mydata[,1:2], 5, nstart = 20)
datacluster
```

```r
datacluster$cluster = as.factor(datacluster$cluster)
plot(x = df$age, y =df$income, xlab = 'age',
     ylab='income', main = 'income with age',
     col = datacluster$cluster, pch=15)
```

```r
library('factoextra')
```

```r
fviz_cluster(datacluster, geom ='point', data=mydata[,1:2]) + ggtitle('k=5')
```

### Comments
- dataset can be clustered effectively based on income group..5 clusters
- age groups are distributed primarily in 2 groups, 20-40 and 40 -60
- correlation between age and income is very low...
